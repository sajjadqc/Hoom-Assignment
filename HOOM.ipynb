```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Change in homeprice from 1963 to 2020 region wise.
home_price = pd.read_csv('Home_prise_region_wise.csv')
home_price.head()
```

```python
home_price.DATE = pd.to_datetime(home_price['DATE'])
home_price.set_index('DATE', inplace = True)
```

```python
home_price.plot.line(rot=0, figsize = (25,15), linewidth=2, fontsize = 15)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Home Price in USD', fontsize=20)
plt.title("USA Home Price (1970-2020)", fontsize=30, color = 'red')
```

## From above graph we can observe that home price is highest in Midwest region and lowest in South region of USA.

```python
# Increase in population from 1990 to 2020 region wise.
population_region = pd.read_csv('Region_population.csv')
population_region.set_index('region', inplace = True)
population_region.plot.bar(rot=0, figsize = (16,8))
plt.ylabel("Population")
plt.title("USA Population 1990-2020", fontsize=25, color = 'red')
```

## From above graph we can observe that there is sharp growth in the population of South and West region as compare to Midwest & Northeast region from 1990 to 2020.

```python
population_region_t = pd.read_csv('pop_transp.csv')
population_region_t.set_index('date', inplace = True)
population_region_t.plot.line(rot=0, figsize = (20,10), linewidth=4, ylabel = 'Population', fontsize = 10)
plt.title("USA Population 1990-2020", fontsize=25, color = 'red')
```

## As per above graph population growth rate will have grate impact on price of home.

```python
HPI = pd.read_csv('demand_data.csv')
HPI.head()
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

norm = normalize(HPI)
```

```python
HPI.set_index('DATE', inplace = True)
```

```python
norm.plot.line(figsize = (20,10), linewidth=4,)
plt.title('Housing price index', fontsize = 25, color = 'red')
```

# Factors influence the home prices across U.S
## As per above graph we can observe following details.
### 1. HPI rises with the decline in un-employment rate.
### 2. HPI rises with the rise in GDP.
### 3. HPI rises with the rise of Mortgage rate.
